```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
data=pd.read_csv('s1_improvised.csv')
data.head()
```

```python
def removeUselessData(data):
    
    data = data.drop('ma_internals',axis=1)
    data = data.drop('ph100_internals',axis=1)
    data = data.drop('be10105_internals',axis=1)
    data = data.drop('be110_internals',axis=1)
    data = data.drop('be103_internals',axis=1)
    data = data.drop('ec100_internals',axis=1)
    data = data.drop('ph110_internals',axis=1)
    data = data.drop('ec110_internals',axis=1)
    data = data.drop('cs110_internals',axis=1)
    
    
#    dataframe = dataframe.drop('G1',axis=1)
#    dataframe = dataframe.drop('G2',axis=1)
    return data
```

```python
s=removeUselessData(data)
```

```python
s.to_csv (r's1_improvised_only_grades.csv', index = False, header=True)
```
